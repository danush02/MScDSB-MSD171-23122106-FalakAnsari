```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load data
df = pd.read_csv('CAC2_Assignment Responses - Form Responses 1.csv')
df
```

```python
# Value counts for categorical vars
df['Stay Location'].value_counts()
df['Which source you depend more for the internet?'].value_counts()
# Bar plot for internet source
source_counts = df['Which source you depend more for the internet?'].value_counts()
ax = source_counts.plot.pie()
```

```python
# Calculate value counts for the 'Where you able to access all required educational content using College Wi-Fi?' 
edu_counts = df['Where you able to access all required educational content using College Wi-F?'].value_counts()
print(edu_counts)
# Create a bar plot to visualize the distribution
ax = edu_counts.plot.bar(rot=0)
ax.set_title('Educational Content Access')
```

```python


# Filter out rows where the "If no, please enter the site names you're not able to access" column is not empty
df[df['If no, please enter the site names you\'re not able to access'].notnull()]

# Count and visualize the top inaccessible sites
top_inaccessible_sites = df['If no, please enter the site names you\'re not able to access'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_inaccessible_sites.values, y=top_inaccessible_sites.index, palette="viridis")
plt.title('Top Inaccessible Sites')
plt.xlabel('Count')
plt.ylabel('Site Names')
plt.show()
```

```python


# Count the number of individuals who were able to access all required educational content using College Wi-Fi
accessed_content_count = df[df['Were you able to access all required educational content using College Wi-Fi?'] == 'Yes'].shape[0]

# Count the number of individuals who were not able to access all required educational content using College Wi-Fi
not_accessed_content_count = df[df['Were you able to access all required educational content using College Wi-Fi?'] == 'No'].shape[0]

# Calculate the percentages
total_count = len(df)
accessed_content_percentage = (accessed_content_count / total_count) * 100
not_accessed_content_percentage = (not_accessed_content_count / total_count) * 100

# Create a pie chart to visualize the percentages
labels = 'Accessed Content', 'Did Not Access Content'
sizes = [accessed_content_percentage, not_accessed_content_percentage]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # Explode the "Accessed Content" slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Add a title
plt.title("Access to Educational Content Using College Wi-Fi")

# Display the pie chart
plt.show()
```
