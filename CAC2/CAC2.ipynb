```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load data
df = pd.read_csv('CAC2_Assignment Responses - Form Responses 1.csv')
df
```

```python
# Value counts for categorical vars
df['Stay Location'].value_counts()
df['Which source you depend more for the internet?'].value_counts()
# Bar plot for internet source
source_counts = df['Which source you depend more for the internet?'].value_counts()
ax = source_counts.plot.pie()
```

```python
# Calculate value counts for the 'Where you able to access all required educational content using College Wi-Fi?' 
edu_counts = df['Where you able to access all required educational content using College Wi-F?'].value_counts()
print(edu_counts)
# Create a bar plot to visualize the distribution
ax = edu_counts.plot.bar(rot=0)
ax.set_title('Educational Content Access')
```

```python


# Filter to responses that provided site names
site_df = df[df['If no, please enter the site names youre not able to access'].notna()]

# Split the site names into separate rows
site_df['If no, please enter the site names youre not able to access'] = site_df['If no, please enter the site names youre not able to access'].str.split(',') 

# Explode the Sites column into separate rows
site_df = site_df.explode('If no, please enter the site names youre not able to access')

# Fill na with unknown
site_df['If no, please enter the site names youre not able to access'].fillna('Unknown', inplace=True) 

# Get value counts of sites
counts = site_df['If no, please enter the site names youre not able to access'].value_counts()

# Print top 10
print(counts.head(10))
# List all column names in your DataFrame
print(df.columns)
```
