```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load data
df = pd.read_csv('CAC2_Assignment_ResponseSheet.csv')
df
```

```python
# Value counts for categorical vars
df['Stay Location'].value_counts()
df['Which source you depend more for the internet?'].value_counts()
# Bar plot for internet source
source_counts = df['Which source you depend more for the internet?'].value_counts()
source_counts.plot.pie(autopct='%1.1f%%', startangle=90)
```

```python
# Calculate value counts for the 'Where you able to access all required educational content using College Wi-Fi?' 
edu_counts = df['Where you able to access all required educational content using College Wi-F?'].value_counts()
print(edu_counts)
# Create a bar plot to visualize the distribution
ax = edu_counts.plot.bar(rot=0,color=['red','green'])
ax.set_title('Educational Content Access')
```

```python


# Count the occurrences of each value in the "Over the past 1 month, how many times have you experienced complete outage of mobile network?" column
outage_counts = df['Over the past 1 month, how many times have you experienced complete outage of mobile network?'].value_counts()

# Sort the counts by index to ensure proper ordering
outage_counts = outage_counts.sort_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
outage_counts.plot(kind='bar', color='skyblue')
plt.title('Mobile Network Outage in the Last Month')
plt.xlabel('Number of Outages')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.show()
```

```python


# Impact on Assignment Submission
plt.figure(figsize=(8, 5))
sns.countplot(df, x='Have network issues ever affected your assignment submission or your preparation for exams?')
plt.title('Impact on Assignment Submission')
plt.show()
```

```python
# 6. Impact on Emergency Communication
plt.figure(figsize=(8, 5))
sns.countplot(df['Have you ever faced difficulties contacting others during emergency situations?'])
plt.title('Impact on Emergency Communication')
plt.show()
```
