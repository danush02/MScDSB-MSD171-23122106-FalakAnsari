```python
class Stack:
    def __init__(self):
        self.stac = []

    def push(self, item):
        self.stac.append(item)

    def pop(self):
        self.stac.pop()

    def print_stack(self):
        if not self.is_empty():
            print("Stack contents:", self.stac)
        else:
            print("Stack is empty")

    def size(self):
        return len(self.stac)

    def top_item(self):
        if not self.is_empty():
            return self.stac[-1]
        else:
            print("Stack is empty")

    def is_empty(self):
        return len(self.stac) == 0


st = Stack()

while True:
    print("\nOptions:")
    print("1. Add Item")
    print("2. Pop Item")
    print("3. Print Stack")
    print("4. Size of Stack")
    print("5. Top Item in the Stack")
    print("6. Check if Stack is Empty")
    print("7. Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        item = input("Enter item to push: ")
        st.push(item)
        print(f"{item} pushed onto the stack.")
    elif choice == "2":
        st.pop()
        print("Item popped from the stack.")
    elif choice == "3":
        st.print_stack()
    elif choice == "4":
        print("Size of stack:", st.size())
    elif choice == "5":
        top_item = st.top_item()
        if top_item is not None:
            print(f"Top item: {top_item}")
    elif choice == "6":
        if st.is_empty():
            print("Stack is empty.")
        else:
            print("Stack is not empty.")
    elif choice == "7":
        break
    else:
        print("Invalid choice. Please try again.")
```
